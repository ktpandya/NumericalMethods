```python
#   Name : Kush Tusharbhai Pandya
#   SR no: 21804
#   Subject : Numerical Methods
#   Department: Aerospace engineering MTech Res
#   Assignment 1 
#   HW1


#       Question 3 (a)
#       Using the Newton’s method, Secant method, and Modified Newton’s method, f ind the solution of f(x) = 0 for the functions listed. 
#       Iterate until you reach a relative tolerance of 10−6 between successive iterates.
#       Report the root found and the number of iterations needed for each method.
#       (a) f(x) = x+e−x2 cosx.
###################################################################################################
#########################Defining the Function#####################################################
import math as mt
from scipy.misc import derivative
def funcT(x):
    return(x+ mt.exp(-x*x)*mt.cos(x))
def funcTD(f,x):
    y1 = derivative(f,x,dx=1)
    return y1    
def funcTDD(f,x):
    y1 = derivative(f,x,dx=1,n=2)
    return y1
print("The function is : x + e^(-x^2)*cos(x)")


#####################################################################################################
########################Input Values for Newton Method###############################################

print("Enter the required tolerance : if it is 10^-n, type n:")
n = int(input())
TOL = 10**(-n)
print(TOL)
print("Enter maximum number of iterations : ")
N = int(input())
print(N)

#####################################################################################################
#################Algorithm for Newtons Method########################################################

def NewtonMethod(f,fd,TOL,N) :
    print("You are using Newtons Method")
    print("Enter initial approximation for Newtons Method :")
    p = int(input())
    print(p)
    for i in range(N):
        q = p - f(p)/fd(f,p)  #Plots the line
   
        if p == 0 and abs((q - p)/q) < TOL:           #Checks for tolerance if initial iteration is a zero, it uses the next iteration to avoid any division by zero error
            print("The root of given equation is ",q," It was obtained after ", i, "iterations")
            break
        elif p!=0 and abs((q - p)/p) < TOL:
            print("The root of given equation is ",q," It was obtained after ", i, "iterations")
            break
        else:
            p = q                           #Switches the values

    i = i+1                             #Next iteration

    print("The value of function at",q ,"is", funcT(q))                     #Output


#####################################################################################################
#########################Algorithm for Secant method#################################################
def SecantMethod(f,fd,TOL,N):
    print("You are using Secant Method : ")
    print("Enter initial two approximations :")      #Input for Secant method
    p0 = int(input())
    p1 = int(input())
    pe = p1
    print(p0,p1)
    
    for i in range(N):
        
        p = p1 - f(p1)*(p1 - p0)/(f(p1) - f(p0))     #Plots the line
                                            #Checks for tolerance
        if abs((p - pe)/pe) < TOL:
            print("The root by secant method is : ",p , "It took ", i , "Iterations to reach it")   #Output

            break
        else:
            pe = p1
            p0 = p1
            p1 = p

    i = i+1
#################################################################################################################
###################Algorithm for Modified Newtons method#########################################################

def ModifiedNewtonMethod(f,fd,fdd,TOL,N):
    print("You are using Modified newtons method")
    print("Enter initial approximation :")
    p0 = int(input())
    print(p0)
    pe = p0
    for i in range(N):

        p = p0 - f(p0)*fd(f,p0)/((fd(f,p0)**2)-f(p0)*fdd(f,p0))
        
        
        if p0 == 0 and abs((p - p0)/p) < TOL:           #Checks for tolerance if initial iteration is a zero, it uses the next iteration to avoid any division by zero error
            print("The root of given equation is ",p," It was obtained after ", i, "iterations for reaching it by modified newtons method")
            break
        elif p0!=0 and abs((p - p0)/p0) < TOL:
            print("The root of given equation is ",p," It was obtained after ", i, "iterations for reaching it by modified newtons method")
            break
        else:

            p0 = p
            pe = p

    i = i + 1 
        

#################################################################################################################

print("Which method you want to use : press 1 for Newtons method, 2 for secant method, 3 for Modified Newtons method")
method = int(input())
if method == 1 :
    NewtonMethod(funcT,funcTD,TOL,N)
elif method == 2 :
    SecantMethod(funcT,funcTD,TOL,N)
elif method == 3 :
    ModifiedNewtonMethod(funcT,funcTD,funcTDD,TOL,N)


 
```
